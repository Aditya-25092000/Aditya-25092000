```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn import model_selection
```

```python
train=pd.read_csv('Downloads/titanic_train.csv')
test=pd.read_csv('Downloads/titanic_test.csv')
```

```python
train.head(20)
```

```python
train_data=train.copy()
```

```python
train_data.columns
```

```python
#handling names
train_data=train_data.drop('Name',axis=1)
#handling sex
train_data.loc[train_data['Sex']=='female','Sex']=int(0)
train_data.loc[train_data['Sex']=='male','Sex']=int(1)
#handling age
train_data.loc[train_data['Age'].isnull(),'Age']=train_data.loc[train_data['Age'].notnull(),'Age'].median()
#handling Embarked
train_data.loc[train_data['Embarked'].isnull() , 'Embarked'] = 'S'
train_data.loc[train_data['Embarked']=='S','Embarked']=int(1)
train_data.loc[train_data['Embarked']=='C','Embarked']=int(2)
train_data.loc[train_data['Embarked']=='Q','Embarked']=int(3)
#handling cabin
train_data=train_data.drop('Cabin',axis=1)
#handling Ticket
train_data=train_data.drop('Ticket',axis=1)
#handling relatives
# train_data['Relatives']=train_data['SibSp']+train_data['Parch']
# train_data.loc[train_data['Relatives']>0,'Relatives']=int(1)
# train_data.loc[train_data['Relatives']==0,'Relatives']=int(0)
# train_data=train_data.drop('SibSp',axis=1)
# train_data=train_data.drop('Parch',axis=1)
# train_data['Survived_or_not']=train_data['Survived']
# train_data=train_data.drop('Survived',axis=1)
```

```python
training_data=train_data.values
x_train=training_data[:,:-1]
y_train=training_data[:,-1]
y_train=y_train.astype('int')
```

```python
test_data=test.copy()
```

```python
test_data.head()
```

```python
#handling names
test_data=test_data.drop('Name',axis=1)
#handling sex
test_data.loc[test_data['Sex']=='female','Sex']=int(0)
test_data.loc[test_data['Sex']=='male','Sex']=int(1)
#handling age
test_data.loc[test_data['Age'].isnull(),'Age']=test_data.loc[test_data['Age'].notnull(),'Age'].mean()
#handling Embarked
test_data.loc[test_data['Embarked'].isnull() , 'Embarked'] = 'S'
test_data.loc[test_data['Embarked']=='S','Embarked']=int(1)
test_data.loc[test_data['Embarked']=='C','Embarked']=int(2)
test_data.loc[test_data['Embarked']=='Q','Embarked']=int(3)
#handling cabin
test_data=test_data.drop('Cabin',axis=1)
#handling Ticket
test_data=test_data.drop('Ticket',axis=1)
#handling relatives
# test_data['Relatives']=test_data['SibSp']+test_data['Parch']
# test_data.loc[test_data['Relatives']>0,'Relatives']=int(1)
# test_data.loc[test_data['Relatives']==0,'Relatives']=int(0)
# test_data=test_data.drop('SibSp',axis=1)
# test_data=test_data.drop('Parch',axis=1)
```

```python
#training data using log. Reg.
clf=LogisticRegression(solver='saga',max_iter=100000,tol=0.00001)
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```

```python
predicted_data=clf.predict(x_test)
```
